##**SINGLE LINE CHART BASED ON PERCENTAGE DROP WITH MNIST AND CIFAR DATASET**

```
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Data
df=pd.DataFrame({'x': [0, 25, 50, 75], 'y1': [3043.3, 7681.27, 8097.95, 8847.90], 'y2': [43.51, 868.91, 1830.82, 4247.47], 'y3': [0.0170, 0.0177, 0.0159, 0.0167] })
 
# multiple line plot
plt.plot( 'x', 'y1', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4,label="Supervised eForest")
plt.plot( 'x', 'y2', data=df, marker='+', color='olive',markersize=12, linewidth=2, label="Unsupervised eForest")
plt.plot( 'x', 'y3', data=df, marker='*', color='red', markersize=12,linewidth=2, linestyle='dashed', label="MLP")
plt.title('MNIST', fontsize=14)
plt.xlabel('DAMAGE IN PERCENT', fontsize=14)
plt.ylabel('MSE', fontsize=14)
plt.grid(True)
plt.legend()
```
